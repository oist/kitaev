```julia
using PyPlot;
```

```julia
a=[cos(π/6),sin(π/6)]
ap=[-cos(π/6),sin(π/6)]
b=[0,1]
```

```julia
#Some Calculations
rows=4;
X=Array{Float64}(2,rows^2)
rows=3
X[:,1]=b
start=1;
nrow=1;
for ii in 1:rows
    top=1;
    for jj in 0:(nrow-1)/2
        top=convert(Int,start+2*jj)
        println(top,' ',top+nrow+1,' ',top+nrow)
        X[:,top+nrow+1]=X[:,top]+b
        X[:,top+nrow]=X[:,top+nrow+1]+a
        println(convert(Int,start+2*jj),' ',convert(Int,start+2*jj)+nrow+1)
    end
    println(' ')
    X[:,top+nrow+2]=X[:,top+nrow+1]+ap
    println(top+nrow+2)
    start+=nrow;
    nrow=2*ii+1;
end
```

```julia
X2=zeros(X)
X2[1,:]=cos(π/3)*X[1,:]-sin(π/3)*X[2,:]
X2[2,:]=sin(π/3)*X[1,:]+cos(π/3)*X[2,:]
```

```julia
PyPlot.svg(true)
pygui(false)
w,h=plt[:figaspect](1)
figure(figsize=(w,h))
scatter(X[1,:],X[2,:],marker="h")
scatter(X[1,:],-X[2,:],marker="h")
scatter(X2[1,:],X2[2,:],marker="h")
scatter(X2[1,:],-X2[2,:],marker="h")
scatter(-X2[1,:],X2[2,:],marker="h")
scatter(-X2[1,:],-X2[2,:],marker="h")
savefig("hexagon3.svg")
```
