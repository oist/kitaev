```julia
function f(kx,ky,Jx,Jy,Jz)
    akx=sqrt(3)*kx/2;
    bky=3*ky/2;
    
    ei=Jx^2+Jy^2+Jz^2
    ei+=2*Jx*Jy*cos(akx*2)
    ei+=2*Jy*Jz*cos(akx-bky)
    ei+=2*Jx*Jz*cos(akx+bky)
    return ei
end

function fprime(kx,ky,Jx,Jy,Jz)
    a=sqrt(3)/2;
    b=3/2;
    akx=a*kx;
    bky=b*ky;
    
    ex=-4*a*Jx*Jy*sin(akx*2);
    ex+=-2*a*Jy*Jz*sin(akx-bky);
    ex+=-2*a*Jx*Jz*sin(akx+bky);
    
    ey=2*b*Jy*Jz*sin(akx-bky);
    ey+=-2*b*Jx*Jz*sin(akx+bky);
    
    return ex, ey 
end
```

```julia
kx=collect((-œÄ-1):.01:(-œÄ+.5));
ky=collect((-1.3):.01:(.3));
kxa=repeat(kx,outer=[1,length(ky)]);
kya=repeat(transpose(ky),outer=[length(kx),1]);
fi=zeros(kxa)
pygui(true)
xlabel("kx")
ylabel("ky")
zlabel("Energy")
#yticks([-œÄ,0,œÄ],["-œÄ","0","œÄ"])
#xticks([-2œÄ,-œÄ,0,œÄ,2œÄ],["-2œÄ","-œÄ","0","œÄ","2œÄ"])
zticks([],[])
title("Eigenvalues of Zero Vortex Lattice")


for kxi in 1:length(kx)
    for kyi in 1:length(ky)
        fi[kxi,kyi]=f(kx[kxi],ky[kyi],1,.5,.5);
    end
end
plot_surface(kxa,kya,fi,cmap="coolwarm",linewidth=0,rstride=1,cstride=1)
plot_surface(kxa,kya,-fi,cmap="coolwarm",linewidth=0,rstride=1,cstride=1)
```
